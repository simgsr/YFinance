```python
import yfinance as yf
import pandas as pd

# Use a real ticker symbol
ticker = '1179.HK'
df = yf.Ticker(ticker)

# Get financial statements
income_stmt = df.incomestmt
balance_sheet = df.balance_sheet
cash_flow = df.cashflow 
company_info = df.info
```

```python
# Combine financial statements
financials_statement = pd.concat([income_stmt, balance_sheet, cash_flow], axis=0)
print(financials_statement.shape)
# display(financials_statement.index.tolist())

# Save to CSV (using the defined ticker variable)
# financials_statement.to_csv(f"{ticker}_combined_data.csv", encoding='utf-8')

# print(financials_statement.index.tolist())
# print(balance_sheet.shape)
# print(cash_flow.shape)
# display(cash_flow)
# cash_flow.to_csv("0087_cash_clow")
# print(sorted(balance_sheet.index.tolist()))
# print(balance_sheet.columns)
# print(balance_sheet.head)
```

```python
# Show all available keys in df.info for reference
print("\n" + "="*80)
print("ALL AVAILABLE INFO KEYS AND VALUES (SORTED):")
print("="*80)
all_keys_sorted = sorted(company_info.keys())
for i, key in enumerate(all_keys_sorted, 1):
    value = company_info.get(key, "N/A")
    print(f"{i:2d}. {key}: {value}")

# Print detailed financial statement data
print("\n" + "="*50)
print("INCOME STATEMENT ITEMS (SORTED):")
print("="*50)
income_stmt = sorted(income_stmt.index.tolist())
for i, item in enumerate(income_stmt, 1):
    print(f"{i:2d}. {item}")

print("\n" + "="*50)
print("BALANCE SHEET ITEMS (SORTED):")
print("="*50)
balance_sheet = sorted(balance_sheet.index.tolist())
for i, item in enumerate(balance_sheet, 1):
    print(f"{i:2d}. {item}")

print("\n" + "="*50)
print("CASH FLOW ITEMS (SORTED):")
print("="*50)
cash_flow = sorted(cash_flow.index.tolist())
for i, item in enumerate(cash_flow, 1):
    print(f"{i:2d}. {item}")

# COMPARISON ANALYSIS
print("\n" + "="*80)
print("OVERLAP ANALYSIS:")
print("="*80)

# Convert to sets for comparison
income_set = set(income_stmt)
balance_set = set(balance_sheet)
cash_flow_set = set(cash_flow)

# Find overlaps
income_balance_overlap = income_set & balance_set
income_cashflow_overlap = income_set & cash_flow_set
balance_cashflow_overlap = balance_set & cash_flow_set
all_three_overlap = income_set & balance_set & cash_flow_set

print(f"\nTotal unique items:")
print(f"  Income Statement: {len(income_set)}")
print(f"  Balance Sheet: {len(balance_set)}")
print(f"  Cash Flow: {len(cash_flow_set)}")

print(f"\nOverlapping items:")
print(f"  Income & Balance: {len(income_balance_overlap)}")
print(f"  Income & Cash Flow: {len(income_cashflow_overlap)}")
print(f"  Balance & Cash Flow: {len(balance_cashflow_overlap)}")
print(f"  All three statements: {len(all_three_overlap)}")

if all_three_overlap:
    print(f"\nItems appearing in ALL three statements:")
    for item in sorted(all_three_overlap):
        print(f"  - {item}")

if income_balance_overlap:
    print(f"\nItems in both Income & Balance (but not Cash Flow):")
    for item in sorted(income_balance_overlap - cash_flow_set):
        print(f"  - {item}")

# Check if financials equals income_stmt
print("\n" + "="*80)
print("VERIFICATION: df.financials vs df.income_stmt")
print("="*80)
print(f"Are they equal? {df.financials.equals(df.income_stmt)}")
```
